## Função Sigmóide

Nesta atividade, vamos iniciar agora um estudo sobre a função sigmóide, também chamada de função logística. A função logística é uma função definida no intervalo de 0 a 1, com a seguinte equação:

$$sigmoid(x) = \frac{1}{1 + e^{-g(x)}}$$

Nesta equação, $g(x)$ pode ser visto como o resultado de uma Regressão Linear. Por ora, vamos assumir que $g(x) = \beta x$, sendo esse $\beta$ uma constante. Como primeiro objetivo, vamos entender como esse valor altera o formato da função logística. 

Primeiramente, vejamos para $\beta = 0.25$

```python
## Biblioteca matemática que usaremos para computar e**x
import math
import matplotlib.pyplot as plt
```

```python
def plot_sigmoid_with_beta(beta) :    
    ## Plotaremos um gráfico com eixo X de -20 a 20.
    eixo_x = range(-20,20,2)
    eixo_y = []

    ## Para cada valor no eixo X, computaremos o valor no eixo Y
    for x in eixo_x :
        eixo_y.append(1/(1 + math.exp(-(beta*x))))
    
    ## Plotando o gráfico    
    plt.plot(eixo_x, eixo_y)
```

```python
plot_sigmoid_with_beta(0.25)
```

Note que esta função possui uma curva em formato de *S*. Vamos aumentar o valor de $\beta$ para ver o que acontece:

```python
## Alterando beta para 0.5
plot_sigmoid_with_beta(0.5)
```

```python
## Alterando beta para 1
plot_sigmoid_with_beta(1)
```

```python
## Alterando beta para 2
plot_sigmoid_with_beta(2)
```

Você já ter notado que alterar $\beta$ altera a suavidade da curva. Em outras palavras, com $\beta = 2$, temos uma transição de 0 para 1 muito mais abrupta do que com $\beta = 0.25$, em que a transição era bem suave e o S da curva bem mais evidente.

Voltando agora para a equação da função sigmóide.

$$sigmoid(x) = \frac{1}{1 + e^{-g(x)}}$$

Vamos assumir que $g(x) = 0.25 x + \gamma$, sendo esse $\gamma$ uma constante, vamos entender como esse valor altera o formato da função logística. 

Primeiramente, vejamos para $\gamma = 1$, e depois alteraremos esse valor:

```python
def plot_sigmoid_with_gamma(gamma) :    
    beta = 0.25
    
    ## Plotaremos um gráfico com eixo X de -20 a 20.
    eixo_x = range(-20,20,2)
    eixo_y = []

    ## Para cada valor no eixo X, computaremos o valor no eixo Y
    for x in eixo_x :
        eixo_y.append(1/(1 + math.exp(-(beta*x + gamma))))
    
    ## Plotando o gráfico    
    plt.plot(eixo_x, eixo_y)
```

```python
plot_sigmoid_with_gamma(1)
```

```python
## Fazendo gamma = 2
plot_sigmoid_with_gamma(2)
```

```python
## Fazendo gamma = 3
plot_sigmoid_with_gamma(3)
```

```python
## Fazendo gamma = -1
plot_sigmoid_with_gamma(-1)
```

```python
## Fazendo gamma = -2
plot_sigmoid_with_gamma(-2)
```

```python
## Fazendo gamma = -3
plot_sigmoid_with_gamma(-3)
```

O que é possível notar é que alterar $\gamma$ causa uma translação na horizontal da curva da sigmóide, mas sem, no entando, mudar o formato da curva em *S*. 

Vamos supor agora que queremos gerar uma translação na vertical e que, além disso, queremos modificar o intervalo para que, ao invés de ser entre 0 e 1, seja entre dois valores quaisquer. Nesse caso, devemos colocar a nossa função sigmóide dentro de uma outra função. Por exemplo:

$f(x) = a*sigmoid(x) + b$

Vamos agora ver o efeito de $a$ em uma sigmóide com $b = 0$, $\beta = 0.25$ e $\gamma = 0$.

```python
def plot_sigmoid_with_a(a) :    
    beta = 0.25
    gamma = 0
    ## Plotaremos um gráfico com eixo X de -20 a 20.
    eixo_x = range(-20,20,2)
    eixo_y = []

    ## Para cada valor no eixo X, computaremos o valor no eixo Y
    for x in eixo_x :
        sigmoid = 1/(1 + math.exp(-(beta*x + gamma)))        
        eixo_y.append( a * sigmoid)
    
    ## Plotando o gráfico    
    plt.plot(eixo_x, eixo_y)
```

```python
## Fazendo a = 3
plot_sigmoid_with_a(3)
```

```python
## Fazendo a = 2
plot_sigmoid_with_a(2)
```

```python
## Fazendo a = 1
plot_sigmoid_with_a(1)
```

```python
## Fazendo a = -1
plot_sigmoid_with_a(-1)
```

```python
## Fazendo a = -2
plot_sigmoid_with_a(-2)
```

O $a$ altera os limites da função de $[0..1]$ para $[0..a]$ no caso de $a>0$ e para $[a..0]$ no caso de $a<0$.

Vamos agora fixar $a = 3$ e entender a alteração causada por $b$

```python
def plot_sigmoid_with_b(b) :    
    beta = 0.25
    gamma = 0
    a = 3
    
    ## Plotaremos um gráfico com eixo X de -20 a 20.
    eixo_x = range(-20,20,2)
    eixo_y = []

    ## Para cada valor no eixo X, computaremos o valor no eixo Y
    for x in eixo_x :
        sigmoid = 1/(1 + math.exp(-(beta*x + gamma)))        
        eixo_y.append( a * sigmoid + b)
    
    ## Plotando o gráfico    
    plt.plot(eixo_x, eixo_y)
```

```python
## Fazendo b = 3
plot_sigmoid_with_b(3)
```

```python
## Fazendo b = 2
plot_sigmoid_with_b(2)
```

```python
## Fazendo b = 1
plot_sigmoid_with_b(1)
```

```python
## Fazendo b = -1
plot_sigmoid_with_b(-1)
```

```python
## Fazendo b = -2
plot_sigmoid_with_b(-2)
```

```python
## Fazendo b = -3
plot_sigmoid_with_b(-3)
```

O que foi possível concluir é que com $a$, $b$, $\beta$ e $\gamma$ temos total controle da curva *S* da função sigmóide. O seu objetivo agora é criar um oráculo com esta curva, configurando os valores manualmente. A seguir, obtemos os valores que você deverá usar para criar o oráculo.

```python
import pandas as pd

faithful = pd.read_csv("../dados/faithful.csv", index_col = 0)
faithful.head()
```

Agora você deverá criar um oráculo propriamente dito, usando a função logística.

```python
import math

## ToDo: meu oráculo
def meu_oraculo(x) :
    gamma = -35
    beta = 0.53
    a     = 2.5
    b     = 2
    sigmoid = 1/(1 + math.exp(-(beta*x + gamma)))
    return a * sigmoid + b
```

```python
## Podemos plotar os valores em um gráfico
## O código a seguir foi visto em aula
def plota_grafico(oraculo, with_error = False) :
    ## Plotando os pontos do banco de dados
    faithful.plot.scatter(x='waiting',
                          y='eruptions')
    
    ## Criando um eixo X para plotar a reta
    eixo_x = range(40,100,2)
    eixo_y = []
    
    ## Gerando a resposta do oráculo
    for x in eixo_x :
        eixo_y.append(oraculo(x))
        
    ## Plotando a reta    
    plt.plot(eixo_x, eixo_y, 'r--', label='Valores Obtidos')
    
    if (with_error) :
    
        ## Colocando barras verticais para evidenciar erros
        ## do oráculo
        pontos_y = faithful.eruptions.values
        pontos_x = faithful.waiting.values
        for i in range(len(pontos_x)) :
            predicted = oraculo(pontos_x[i])
            plt.plot(
                (pontos_x[i], pontos_x[i]),
                (predicted, pontos_y[i]),
                '-', c = "black"
            )
```

```python
plota_grafico(meu_oraculo)
```
